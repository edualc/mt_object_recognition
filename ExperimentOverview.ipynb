```python
df_a = pd.read_csv('wandb_vggonly.csv')
df_b = pd.read_csv('wandb_vgglcl.csv')
df_c = pd.read_csv('wandb_vgg16lcl.csv')

df_c.head()
```

```python
df_both = pd.merge(pd.merge(df_a, df_b), df_c)
model_names = sorted(list(set([x.split(' - ')[0] for x in df_both.columns])))[:-1]
model_names
```

```python
data = dict()
for m in model_names:
    tmp = {
        'iteration': np.array(df_both['iteration']),
        'val_acc': np.array(df_both[m + ' - val_acc']),
        'val_loss': np.array(df_both[m + ' - val_loss']),
        'test_acc': np.array(df_both[m + ' - test_acc']),
        'test_loss': np.array(df_both[m + ' - test_loss'])
    }
    data[m] = tmp
    
df = pd.DataFrame(data).transpose()
df['LCL'] = ['LCL' in x for x in df.index]

vgg_experiment = [ 'VGG19_2022-04-04_160910', 'VGG19_2022-04-04_164603', 'VGG19_2022-04-04_172253', 'VGG19_2022-04-04_175945', 'VGG19_2022-04-04_183636', 'VGG19_2022-04-04_191333' ]
iota_experiment = [ 'VGG19_LCL_2022-04-05_124057', 'VGG19_LCL_2022-04-05_124104', 'VGG19_LCL_2022-04-05_171610', 'VGG19_LCL_2022-04-05_182033', 'VGG19_LCL_2022-04-05_212941', 'VGG19_LCL_2022-04-06_001414']
eta_experiment = [ 'VGG19_LCL_2022-04-05_155542', 'VGG19_LCL_2022-04-05_155601', 'VGG19_LCL_2022-04-05_205611', 'VGG19_LCL_2022-04-05_214333', 'VGG19_LCL_2022-04-06_015226', 'VGG19_LCL_2022-04-06_023811' ]
experiment = []
for ix in df.index:
    if ix in iota_experiment:
        experiment.append('iota')
    elif ix in eta_experiment:
        experiment.append('eta')
    elif ix in vgg_experiment:
        experiment.append('vgg_only')
    else:
        experiment.append('vgg16_lcl')
df['experiment'] = experiment
df
```

```python
for index, row in df.iterrows():
    best = 1000
    counter = 0
    best_index = [-1, -1, -1, -1]
    
    for i, val_loss in enumerate(row['val_loss']):
        
        if val_loss < best:
            best = val_loss
            counter = 0
        else:
            counter += 1

            if counter == 1:
                if best_index[0] < 0:
                    best_index[0] = i-1
                
            if counter == 2:
                if best_index[1] < 0:
                    best_index[1] = i-1
            
            if counter == 3:
                if best_index[2] < 0:
                    best_index[2] = i-1
                    
    for k, ix in enumerate(best_index):
        if ix == -1:
            best_index[k] = i
            
    df.loc[index, 'early_stop_1'] = best_index[0]
    df.loc[index, 'early_stop_2'] = best_index[1]
    df.loc[index, 'early_stop_3'] = best_index[2]
    df.loc[index, 'early_stop_4'] = best_index[2]
    
    best = 0
    counter = 0
    best_index = [-1, -1, -1, -1]
    
    for i, val_acc in enumerate(row['val_acc']):
        
        if val_acc > best:
            best = val_acc
            counter = 0
        else:
            counter += 1

            if counter == 1:
                if best_index[0] < 0:
                    best_index[0] = i-1
                
            if counter == 2:
                if best_index[1] < 0:
                    best_index[1] = i-1
            
            if counter == 3:
                if best_index[2] < 0:
                    best_index[2] = i-1
                    
    for k, ix in enumerate(best_index):
        if ix == -1:
            best_index[k] = i
            
    df.loc[index, 'early_stop_1__acc'] = best_index[0]
    df.loc[index, 'early_stop_2__acc'] = best_index[1]
    df.loc[index, 'early_stop_3__acc'] = best_index[2]
    df.loc[index, 'early_stop_4__acc'] = best_index[2]
```

```python
import pandas as pd
df_lcl = pd.read_csv('mnist_c__lcl.csv')
df_lcl['experiment'] = 'VGG19+LCL'

df_vgg = pd.read_csv('mnist_c__vggonly.csv')
df_vgg['experiment'] = 'VGG19'

df_vgg16_1 = pd.read_csv('mnist_c__vgg16_lcl01.csv')
df_vgg16_1['experiment'] = 'VGG16+LCL'
df_vgg16_2 = pd.read_csv('mnist_c__vgg16_lcl23.csv')
df_vgg16_2['experiment'] = 'VGG16+LCL'
df_vgg16_3 = pd.read_csv('mnist_c__vgg16_lcl45.csv')
df_vgg16_3['experiment'] = 'VGG16+LCL'


df = df_lcl.append(df_vgg)
df = df.append(df_vgg16_1)
df = df.append(df_vgg16_2)
df = df.append(df_vgg16_3).reset_index()
df = df[['model','mnist_c_variant','accuracy','experiment']]
print(df.groupby(['experiment']).count())
df
```

```python
df_mean = df.groupby(['mnist_c_variant', 'experiment']).mean()['accuracy']
df_std = df.groupby(['mnist_c_variant', 'experiment']).std()['accuracy']

def filter_vgg_lcl(df, attr):
    df_x = df.reset_index()
    df_x = df_x[df_x['experiment'] == 'VGG19+LCL'][['mnist_c_variant','accuracy']]
    df_x.columns = ['mnist_c', attr+'_lcl']
    df_x

    df_y = df.reset_index()
    df_y = df_y[df_y['experiment'] == 'VGG19'][['mnist_c_variant','accuracy']]
    df_y.columns = ['mnist_c', attr+'_vgg']
    df_y

    df_z = df.reset_index()
    df_z = df_z[df_z['experiment'] == 'VGG16+LCL'][['mnist_c_variant','accuracy']]
    df_z.columns = ['mnist_c', attr+'_lcl16']
    df_z
    
    return df_x, df_y, df_z
    
a,b,c = filter_vgg_lcl(df_mean, 'accuracy')
d,e,f = filter_vgg_lcl(df_std, 'std')



dfx = pd.merge(pd.merge(pd.merge(a,b), pd.merge(c,d)),pd.merge(e,f))
dfx
```

```python
a = pd.read_csv('mnist_c__vgg19r_lcl534.csv')
a['experiment'] = 'lcl@5'
b = pd.read_csv('mnist_c__vgg19r_lcl5012.csv')
b['experiment'] = 'lcl@5'
df_lclr = a.append(b)

a = pd.read_csv('mnist_c__vgg19r_lcl5__es3012.csv')
a['experiment'] = 'lcl@5_es3'
b = pd.read_csv('mnist_c__vgg19r_lcl5__es334.csv')
b['experiment'] = 'lcl@5_es3'
df_lclr = df_lclr.append(a.append(b))

a = pd.read_csv('mnist_c__vggonly012.csv')
a['experiment'] = 'vggonly'
b = pd.read_csv('mnist_c__vggonly345.csv')
b['experiment'] = 'vggonly'
df_lclr = df_lclr.append(a.append(b))

a = pd.read_csv('experiment__vgg19_vs_vgg19LCL/mnist_c__vggonly.csv')
a['experiment'] = 'vggonly_es3'
df_lclr = df_lclr.append(a)
df_lclr
```

```python
df_mean = df_lclr.groupby(['mnist_c_variant', 'experiment']).mean()['accuracy']
df_std = df_lclr.groupby(['mnist_c_variant', 'experiment']).std()['accuracy']

def filter_vggr(df, attr):
    df_x = df.reset_index()
    df_x = df_x[df_x['experiment'] == 'lcl@5'][['mnist_c_variant','accuracy']]
    df_x.columns = ['mnist_c', attr+'_lcl@5']
    df_x

    df_y = df.reset_index()
    df_y = df_y[df_y['experiment'] == 'lcl@5_es3'][['mnist_c_variant','accuracy']]
    df_y.columns = ['mnist_c', attr+'_lcl@5_es3']
    df_y

    df_z = df.reset_index()
    df_z = df_z[df_z['experiment'] == 'vggonly'][['mnist_c_variant','accuracy']]
    df_z.columns = ['mnist_c', attr+'_vggonly']
    df_z

    df_w = df.reset_index()
    df_w = df_w[df_w['experiment'] == 'vggonly_es3'][['mnist_c_variant','accuracy']]
    df_w.columns = ['mnist_c', attr+'_vggonly_es3']
    df_w
    
    return df_x, df_y, df_z, df_w
    
a,b,c,d = filter_vggr(df_mean, 'accuracy')
e,f,g,h = filter_vggr(df_std, 'std')

dfx = pd.merge(pd.merge(pd.merge(pd.merge(a,b), pd.merge(c,d)),pd.merge(e,f)), pd.merge(g,h))
dfx
```
